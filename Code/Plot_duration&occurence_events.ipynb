```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read Excel file
Duration_Events = pd.read_excel(r"C:\Users\raulo\Desktop\Faculdade\Mestrado Bioestatística\Tese\Escrita\Book1.xlsx", sheet_name="Duration_Events")

Occurence_Events = pd.read_excel(r"C:\Users\raulo\Desktop\Faculdade\Mestrado Bioestatística\Tese\Escrita\Book1.xlsx", sheet_name="Occurence_Events")
```

```python
Duration_Events
```

```python
# Extract event names and corresponding time values
events = Duration_Events['Event']
time_values = Duration_Events['time']

# Plot the bar chart
plt.figure(figsize=(7, 5))
plt.bar(events, time_values, color='blue', alpha=0.7, width=0.3)
plt.title('')
plt.xlabel('')
plt.ylabel('Duration (seconds)')
plt.xticks(rotation=0, ha='center')  # Rotate x-axis labels for better readability
plt.tight_layout()

plt.savefig(r"C:\Users\raulo\Desktop\Faculdade\Mestrado Bioestatística\Tese\Escrita\imagens\Duration_Events.png")

plt.show()
```

```python
# Plot the bar chart
plt.figure(figsize=(7, 5))
plt.bar(Occurence_Events['Event'], Occurence_Events['Occurence'], color='blue', alpha=0.7, width=0.3)
plt.title('')
plt.xlabel('')
plt.ylabel('Occurences ')
plt.xticks(rotation=45, ha='center')  # Rotate x-axis labels for better readability
plt.tight_layout()

plt.savefig(r"C:\Users\raulo\Desktop\Faculdade\Mestrado Bioestatística\Tese\Escrita\imagens\Duration_Events.png")

plt.show()
```

```python
# Plot histograms side by side
plt.figure(figsize=(10, 5))

# Plot the bar chart
plt.subplot(1, 2, 1)
plt.bar(Duration_Events['Event'], Duration_Events['time'], color='blue', alpha=0.7, width=0.3)
plt.title('Events Duration')
plt.xlabel('')
plt.ylabel('Duration (seconds)')
plt.xticks(rotation=45, ha='center')  # Rotate x-axis labels for better readability




plt.subplot(1, 2, 2)
plt.bar(Occurence_Events['Event'], Occurence_Events['Occurence'], color='blue', alpha=0.7, width=0.3)
plt.title('Events Occurence')
plt.xlabel('')
plt.ylabel('Occurences ')
plt.xticks(rotation=45, ha='center')  # Rotate x-axis labels for better readability
plt.tight_layout()

plt.savefig(r"C:\Users\raulo\Desktop\Faculdade\Mestrado Bioestatística\Tese\Escrita\imagens\Duration&Occurences_Events.png")

plt.show()
```
