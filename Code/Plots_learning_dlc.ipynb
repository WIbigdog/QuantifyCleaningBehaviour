```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read Excel file
Front_learning = pd.read_csv(r"C:\Users\raulo\Desktop\Faculdade\Mestrado Bioestatística\Tese\Escrita\DLC\learning_stats.csv", sep=";", decimal=",")

Top_learning = pd.read_csv(r"C:\Users\raulo\Desktop\Faculdade\Mestrado Bioestatística\Tese\Escrita\DLC\top_learning_stats.csv", sep=";", decimal=",")
```

```python
# Plot line plot
plt.plot(Front_learning["iteration"], Front_learning["loss"], color='blue', label='Loss Line')

# Add labels and title
plt.xlabel('Iteration')
plt.ylabel('Loss Value')
plt.title('')

# Set x-axis ticks at intervals of every 15000
plt.xticks(range(0, Front_learning["iteration"].max() + 1, 15000))

plt.savefig(r"C:\Users\raulo\Desktop\Faculdade\Mestrado Bioestatística\Tese\Escrita\imagens\Front_learning.png")

# Show plot
plt.show()
```

```python
# Plot line plot
plt.plot(Top_learning["iteration"], Top_learning["loss"], color='blue', label='Loss Line')

# Add labels and title
plt.xlabel('Iteration')
plt.ylabel('Loss Value')
plt.title('')

# Set x-axis ticks at intervals of every 15000
plt.xticks(range(0, Top_learning["iteration"].max() + 1, 5000))

plt.savefig(r"C:\Users\raulo\Desktop\Faculdade\Mestrado Bioestatística\Tese\Escrita\imagens\Top_learning.png")

# Show plot
plt.show()
```
