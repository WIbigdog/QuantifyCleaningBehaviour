```python
import pandas as pd
import os
```

```python
directory = r'C:\Users\raulo\Desktop\ze_vids\phase_3\results_dataframes\CleanerCentroid4_Client5_smaller'
for filename in os.listdir(directory):
    if (filename[:4] == "LD04") & filename.endswith('.csv'):
        df = pd.read_csv(os.path.join(directory, filename))
        # Initialize a list to store lists of continuous "interaction" sequences
        interaction_sequences = []

        final_count = 0

        # Initialize variables to track the current sequence
        current_sequence = []
        previous_behavior = None

        # Iterate over the rows of the DataFrame
        for index, row in df.iterrows():
            # Check if the current behavior is "interaction"
            if row['Behavior'] == 'interaction':
                # If previous behavior was not "interaction" or it's the first row, start a new sequence
                if previous_behavior != 'interaction' or previous_behavior is None:
                    current_sequence = [(index, row['Interaction_Predictions'])]
                else:
                    # If the previous behavior was "interaction", append the index and value to the current sequence
                    current_sequence.append((index, row['Interaction_Predictions']))
            else:
                # If the behavior is not "interaction", check if there was a previous sequence and store it
                if current_sequence:
                    interaction_sequences.append(current_sequence)
                    current_sequence = []
            previous_behavior = row['Behavior']

        # Check if there's a remaining sequence at the end of the DataFrame
        if current_sequence:
            interaction_sequences.append(current_sequence)

        # Calculate the number of occurrences of grouped events of "interaction" within each sequence
        grouped_events_per_sequence = []
        for sequence in interaction_sequences:
            grouped_event_count = 0
            interaction_seen = False
            for _, value in sequence:
                if value == "interaction":
                    if not interaction_seen:
                        grouped_event_count += 1
                    interaction_seen = True
                else:
                    interaction_seen = False
            grouped_events_per_sequence.append(grouped_event_count)

        # Print the number of occurrences of grouped events of "interaction" within each sequence
        for i, count in enumerate(grouped_events_per_sequence, start=1):
            #print(f"Sequence {i} - Number of grouped events of 'interaction': {count}")

            if count >=2:
                final_count += count -1

        print(f"{filename}Final count is {final_count}")
```
